# Question (4) (d):
## In a graph with logarithmic scale, plot N vs E(N ) for rd = 1/2 and 1/6. Obtain the order of accuracy for the two rd cases.

# Answer (4) (d):

# Reading the average of absolute error from the files

```python
rd = []
for i in range(2,len(Output_files),3):
    temp = pd.read_csv(Output_files[i], delimiter = ",",header=None).to_numpy().squeeze()
    rd.append(temp.item())
rd_half = []
for i in range(0,len(rd),2):
    temp = rd[i]
    rd_half.append(temp)
rd_one_sixth = []
for i in range(1,len(rd),2):
    temp = rd[i]
    rd_one_sixth.append(temp)
```

## Average of absolute error for rd = 1/2

```python
rd_half
```

## Average of absolute error for rd = 1/6

```python
rd_one_sixth
```

## Values of N

```python
N = [32,64,128,256]
```

# Average error vs step size for different values of rd in log-log scale

```python
fig = plt.figure()
plt.loglog(N,rd_half)
plt.loglog(N,rd_one_sixth)
plt.plot(N,(5.5*np.array(N))**(-3),"--")
plt.plot(N,(0.5*np.array(N))**(-5),"--")
plt.plot(N,(0.70*np.array(N))**(-2),"--")
plt.legend(["r_d = 1/2", "r_d = 1/6", "Slope = -3","Slope = -5","Slope = -2"])
plt.title("N vs E(N)")
plt.xlabel("Step size (N)")
plt.ylabel("Average error (E(N))")
#plt.xticks(ticks=Step_size_num,rotation=90)
plt.grid(which='both')
plt.show()
fig.savefig("E_N_vs_N.png",dpi = 500, bbox_inches="tight")
```
