## Plotting the results for rd = 1/2 and N = 32 at $t_{end} = 0.4$

# Average error vs step size for different values of rd in log-log scale

```python
rd = []
for i in range(2,len(Output_files),3):
    temp = pd.read_csv(Output_files[i], delimiter = ",",header=None).to_numpy().squeeze()
    rd.append(temp.item())
rd_half = []
for i in range(0,len(rd),2):
    temp = rd[i]
    rd_half.append(temp)
rd_one_sixth = []
for i in range(1,len(rd),2):
    temp = rd[i]
    rd_one_sixth.append(temp)
```

```python
rd
```

```python
rd_half
```

```python
rd_one_sixth
```

```python
N = [32,64,128,256]
```

```python
fig = plt.figure()
plt.loglog(N,rd_half)
plt.loglog(N,rd_one_sixth)
plt.plot(N,(1.50*np.array(N))**(-2),"--")
plt.legend(["r_d = 1/2", "r_d = 1/6", "Slope = -2"])
plt.xlabel("N")
plt.ylabel("Average error")
plt.grid(which='both')
plt.show()
fig.savefig("Average_Error_log.png",dpi = 500, bbox_inches="tight")
```

```python
Q_4_Error_Data = pd.read_csv("Question_4_Error_Data.csv",header=None)
```

# Average error vs step size for different methods in log-log scale

```python
fig = plt.figure()
plt.loglog(N,rd_half)
plt.loglog(N,Q_4_Error_Data)
plt.plot(N,(1.50*np.array(N))**(-2),"--")
plt.legend(["Implicit", "Explicit", "Slope = -2"])
plt.xlabel("N")
plt.ylabel("Average of the absolute error")
plt.grid(which='both')
plt.title(R"Tolerance = $10^{-6}$ ($r_{d}$ = 1/2)")
plt.show()
fig.savefig("Average_Error_log_for_Different_Methods.png",dpi = 500, bbox_inches="tight")
```

## In case of the implicit method, the truncation error dominates, when we are using lesser number of grid points but when we are using a finer grid the propagation error keep on accumulating over iterations and eventually dominates the truncation error. Hence, the error increases on using a finer grid, for a fixed tolerance.
## We have found that on decreasing the tolerance from the $10^{-4}$ to $10^{-6}$ the error keeps on reducing.
